```python
features=[]
labels=[]
training_length=70
for seq in sequences:
    for i in range(training_length,len(seq)):
        extract=seq[i-training_length:i+1]
        features.append(extract[:-1])
        labels.append(extract[-1])
```

```python
from sklearn.model_selection import train_test_split
```

```python
sequences=sequences.reshape(7,390,71)
```

```python
comp2=label_array.reshape(7,390,8)
```

```python
x_train,x_test=train_test_split(sequences,test_size=1/7,shuffle=False)
```

```python
y_train,y_test=train_test_split(comp2,test_size=1/7,shuffle=False)
```

```python
y_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense,Dropout
```

```python
model=Sequential()
```

```python
#recurrent layer
model.add(LSTM(64,input_shape=(x_train.shape[1:]),activation='tanh',return_sequences=True))
```

```python
model.add(Dropout(0.2))
```

```python
#fully connected layer
model.add(Dense(64,activation='sigmoid'))
```

```python
#dropout for regularization
model.add(Dropout(0.2))
```

```python
#output layer
model.add(Dense(num_words,activation='softmax'))
```

```python
model.summary()
```

```python
opt=tf.keras.optimizers.Adam(lr=1e-3 ,decay=1e-5)
```

```python
from tensorflow.keras import backend as K
accuracy=tf.keras.metrics.Accuracy()
cce=tf.keras.losses.CategoricalCrossentropy()
```

```python
def f1_score(y_true, y_pred): 
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    recall = true_positives / (possible_positives + K.epsilon())
    f1_val = 2*(precision*recall)/(precision+recall+K.epsilon())
    return f1_val
```

```python
#compile the model
model.compile(optimizer=opt,loss=cce,metrics=['accuracy'])
```

```python
history=model.fit(x_train,y_train,
                  epochs=5,validation_data=(x_test,y_test))
```

```python
mutation,mutation_accuracy=model.evaluate(x_test,y_test)
```

```python
print('Mutation: {0:.2f}. Mutation accuracy: {1:.2f}%'.format(mutation*100, mutation_accuracy*100.))
```
