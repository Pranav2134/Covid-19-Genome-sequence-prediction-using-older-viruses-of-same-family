```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
df=pd.read_csv("complete_timeline_data.csv")
```

```python
df
```

```python
df=df.dropna()
```

```python
df
```

```python
abstracts=df['sequence']
```

```python
(abstracts[1][:426])
```

```python
tokenizer=tf.keras.preprocessing.text.Tokenizer(num_words=None,filters='\n',lower=False,char_level=True)
```

```python
tokenizer.fit_on_texts(abstracts)
```

```python
sequences=tokenizer.texts_to_sequences(abstracts)
```

```python
sequences[10][:15]
```

```python
df['character_token']=sequences
```

```python
sequences=np.array(sequences)
```

```python
sequences.shape
```

```python
idx_word=tokenizer.index_word
' '.join(idx_word[w] for w in sequences[100][:40])
```

```python
features=[]
labels=[]
training_length=20
for seq in sequences:
    for i in range(training_length,len(seq)):
        extract=seq[i-training_length:i+1]
        features.append(extract[:-1])
        labels.append(extract[-1])
```

```python
features
```

```python
#features=np.array(features)
```

```python
num_words=len(idx_word)+1
num_words
```

```python
label_array=np.zeros((len(features),num_words),dtype=np.int8)
```

```python
for example_index,word_index in enumerate(labels):
    label_array[example_index,word_index]=1
label_array.shape
```

```python
label_array
```

```python
idx_word[np.argmax(label_array)]
```

```python
sequences.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
sequences=sequences.reshape(7,390,71)
```

```python
x_train,x_test=train_test_split(sequences,test_size=1/7,shuffle=False)
```

```python
x_train,x_valid=train_test_split(x_train,test_size=1/6,shuffle=False)
```

```python
x_valid.shape
```

```python
x_train[0].shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense,Dropout
```

```python
model=Sequential()
```

```python
#recurrent layer
model.add(LSTM(64,input_shape=(390,71),activation='relu',return_sequences=False))
```

```python
model.add(Dropout(0.3))
```

```python
#fully connected layer
model.add(Dense(16,activation='relu'))
```

```python
#dropout for regularization
model.add(Dropout(0.5))
```

```python
#output layer
model.add(Dense(num_words,activation='softmax'))
```

```python
opt=tf.keras.optimizers.Adam(lr=1e-3 ,decay=1e-5)
```

```python
#compile the model
model.compile(optimizer=opt,loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(x_train,
                  epochs=5,
                  validation_data=x_valid
                 )
```
