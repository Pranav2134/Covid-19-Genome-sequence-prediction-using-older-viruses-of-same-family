```python
import numpy as np
import pandas as pd
```

```python
file=open("MERS-Cov.txt","r")
line_no=0
data=[[]]
```

```python
for line in file:
    data.append([line_no,line])
    line_no+=1
```

```python
line_no
```

```python
df=pd.DataFrame(data,columns=['line','sequence'])
```

```python
df_mod=df.copy()
df_mod
```

```python
df_mod=df_mod.drop(df_mod.index[0],axis=0)
```

```python
df_mod
```

```python
df_mod_saved=df_mod.copy()
```

```python
df_mod_saved
```

```python
df_mod_saved.to_csv(r'betaCov2013-sequence.csv', index = False)
```
